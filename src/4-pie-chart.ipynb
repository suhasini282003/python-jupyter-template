### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.

import matplotlib.pyplot as plt

def validate_pie_data(labels, sizes):
    """Function to validate pie chart input data before plotting."""
    if not isinstance(labels, list) or not isinstance(sizes, list):
        raise ValueError("labels and sizes should be lists.")
    
    if len(labels) == 0 or len(sizes) == 0:
        raise ValueError("labels and sizes cannot be empty.")
    
    if len(labels) != len(sizes):
        raise ValueError("labels and sizes must have the same length.")
    
    if not all(isinstance(size, (int, float)) and size >= 0 for size in sizes):
        raise ValueError("All elements in sizes should be non-negative numbers.")

# Pie chart data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Validate pie chart data
validate_pie_data(labels, sizes)

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'])

# Title
plt.title('Programming Language Popularity')

# Show theÂ plot
plt.show()
```
