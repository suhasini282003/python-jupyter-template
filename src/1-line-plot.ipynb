### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

def validate_data(x, y):
    """Function to validate input data before plotting."""
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("Both x and y should be lists.")
    
    if len(x) == 0 or len(y) == 0:
        raise ValueError("x and y cannot be empty.")
    
    if len(x) != len(y):
        raise ValueError("x and y must have the same length.")
    
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise ValueError("All elements in x and y should be numeric (int or float).")

# Data
x = list(range(11))  # x values from 0 to 10
y = [i**2 for i in x]  # Squaring each x value

# Validate data
validate_data(x, y)

# Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customizing the plot
plt.xlabel('X Values')  # X-axis label
plt.ylabel('Y Values (Squared)')  # Y-axis label
plt.title('Line Plot of y = x^2')  # Title
plt.grid(True, linestyle='--', alpha=0.7)  # Adding a grid
plt.legend()  # Show legend

# Display the plot
plt.show()
```
