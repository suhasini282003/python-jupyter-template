### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

def validate_histogram_data(data):
    """Function to validate input data for the histogram."""
    if not isinstance(data, list) or len(data) == 0:
        raise ValueError("data should be a non-empty list.")
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError("All elements in data should be numeric (int or float).")

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Validate histogram data
validate_histogram_data(data)

# Create a histogram
plt.figure(figsize=(6, 4))
plt.hist(data, bins=5, color='blue', edgecolor='black')

# Customizing the histogram
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the histogram
plt.show()
```
