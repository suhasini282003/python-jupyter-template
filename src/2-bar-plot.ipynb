### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.

import matplotlib.pyplot as plt

def validate_data(categories, values, colors):
    """Function to validate input data before plotting."""
    if not isinstance(categories, list) or not isinstance(values, list) or not isinstance(colors, list):
        raise ValueError("categories, values, and colors should all be lists.")
    
    if len(categories) == 0 or len(values) == 0 or len(colors) == 0:
        raise ValueError("categories, values, and colors cannot be empty.")
    
    if len(categories) != len(values) or len(categories) != len(colors):
        raise ValueError("categories, values, and colors must have the same length.")
    
    if not all(isinstance(cat, str) for cat in categories):
        raise ValueError("All elements in categories should be strings.")
    
    if not all(isinstance(val, (int, float)) for val in values):
        raise ValueError("All elements in values should be numeric (int or float).")
    
    if not all(isinstance(col, str) for col in colors):
        raise ValueError("All elements in colors should be strings representing valid colors.")

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Validate data
validate_data(categories, values, colors)

# Create a bar plot
plt.bar(categories, values, color=colors)

# Customizing the plot
plt.xlabel('Categories')  # X-axis label
plt.ylabel('Values')  # Y-axis label
plt.title('Bar Plot with Different Colors')  # Title
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding a grid on Y-axis

# Display the plot
plt.show()
```
